## Exercise 1: 
Create an abstract class called Vehicle with 
- color, make, and year private attributes 

and abstract method called 
- get_info() which returns all information
- honk() which returns a string of honking sound

Create Car, Boat, and Bicycle class which is a child class of the Vehicle class.  Then, create an instance of each child class and call get_info() method to print out its information in a user friendly format. Add other methods as you like.

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
    @abstractmethod
    def get_info(self):
        pass
    def honk(self):
        pass

class Car(Vehicle):
    def __init__(self, color, make, year):
        super().__init__(color, make, year)
        self.__color = color
        self.__make = make
        self.__year = year

    def get_info(self):
        return f"Color: {self.__color} \n Make: {self.__make} \n Year: {self.__year}"

    def honk(self):
        return "BEEP"

class Boat(Vehicle):
    def __init__(self, color, make, year, floats):
        super().__init__(color, make, year)
        self.__color = color
        self.__make = make
        self.__year = year
        self.__float = floats

    def get_info(self):
        return f"Color: {self.__color} \n Make: {self.__make} \n Year: {self.__year}\n Floats: {self.__float} \n"

    def honk(self):
        return f"HONK"


class Bicycle(Vehicle):
    def __init__(self, color, make, year, wheels):
        super().__init__(color, make, year)
        self.__color = color
        self.__make = make
        self.__year = year
        self.__wheels = wheels

    def get_info(self):
        return f"Color: {self.__color} \n Make: {self.__make} \n Year: {self.__year} \n Wheels: {self.__wheels} \n"

    def honk(self):
        return "Beep"

c1 = Car("red", "prius", 2000)
b1 = Boat("Red", "speedboat", 2001, "Yes")
Bi1 = Bicycle("Blue", "Mountain", 2020, 2)

print(c1.get_info())
print(b1.get_info())
print(Bi1.get_info())
print(c1.honk())
print(b1.honk())
print(Bi1.honk())
```

## Exercise 2: 
Create an abstract class called Person with name and address attributes, and print_info() abstract method. 
1. Build an Employee class inherited from Person class that 
    - stores hired date and salary attributes 
    - print_info() method that will be used to print out all information about an employee
    - a method to calculate employee's monthly pay where federal tax is 15%, MD tax is 4.5%, and a local tax is 3%.
    
2. Build a Customer class also inherited from Person class that
    - stores balance attribute
    - print_info() method that will be used to print out all information about a customer


Build any method you need but apply encapsulation at all possible. Then, write a program to show your classes work with multiple employees and customers. 

CHALLENGE!!!
Create an Address class and use it in Person class

```python
from abc import ABC, abstractmethod
class Person(ABC):
    def __init__(self, name, address):
        self.name = name
        self.address = address

    @abstractmethod
    def print_info(self):
        pass


class Employee(Person):
    def __init__(self, name, address, hired, salary):
        super().__init__(name, address)
        self.__hired = hired
        self.__salary = float(salary)

    def print_info(self):
        return f"{self.name} from {self.address} was hired in {self.__hired} and makes ${self.__salary}"

    def pay(self):
        pay = self.__salary * .85 * .955 * .97
        pay = round(pay, 2)
        return f"With tax the salary is: {pay}"


class Customer(Person):
    def __init__(self, name, address, balance):
        super().__init__(name, address)
        self.__balance = balance

    def print_info(self):
        return f"{self.name} from {self.address} has ${self.__balance} in their balance."



E1 = Employee("John", "1st street", 2000, "53000")
E2 = Employee("Boss", "boss street", 2020, "100000000")
print(E1.print_info())
print(E1.pay())
print(E2.print_info())
print(E2.pay())

c1 = Customer("Billy", "2nd street", "1")
c2 = Customer("Bob", "8th place", "100000")
print(c2.print_info())
print(c1.print_info())
```
